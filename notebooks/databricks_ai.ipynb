```python
import requests
import json
from IPython.display import display, Markdown
# databricks-gpt-oss-20b

def call_databricks_claude_model_api(prompt: str, model_name: str = "databricks-claude-opus-4-8"):
    """
    Calls a Databricks Serving Endpoint using standard HTTP requests.
    This method is useful when you cannot use the Databricks SDK.
    """

    # 1. AUTHENTICATION & CONFIGURATION
    # Best Practice: dbutils is the safest way to grab context-aware credentials
    # within a notebook environment without hardcoding tokens.
    context = dbutils.notebook.entry_point.getDbutils().notebook().getContext()
    host = context.apiUrl().get()
    token = context.apiToken().get()
    
    # Construct the REST API URL for the specific model
    url = f"{host}/ai-gateway/mlflow/v1/chat/completions"
    
    # 2. PREPARE THE PAYLOAD
    headers = {
        "Authorization": f"Bearer {token}", 
        "Content-Type": "application/json"
    }
    
    data = {
        "model": f"{model_name}",
        "max_tokens": 1024,
        "messages": [
            {"role": "system", "content": "You are a concise expert."},
            {"role": "user", "content": prompt}
        ],
        "temperature": 0.5
    }

    # 3. EXECUTE THE REQUEST
    try:
        response = requests.post(url, headers=headers, json=data)
        
        # Check if the request was successful (HTTP 200)
        response.raise_for_status()
        
        # Parse the JSON response
        response_json = response.json()
        
        # 4. DATA EXTRACTION
        # The response structure can vary slightly depending on the model,
        # so we use a robust extraction method.
        content = response_json['choices'][0]['message']['content']
        
        if isinstance(content, list):
            # Handle cases where the model returns structured text blocks
            output = "".join([p['text'] for p in content if p.get('type') == 'text'])
        else:
            output = content

        return output

    except requests.exceptions.RequestException as e:
        return f"API Request failed: {e}"
```

```python
# --- ENVIRONMENT SETUP ---
# It's often better to run installs in a separate cell at the very top.
# Using 'quiet' mode (-q) keeps the notebook output clean for new users.
%pip install -q databricks-langchain
dbutils.library.restartPython()

# --- IMPORTS ---
import os
from databricks_langchain import ChatDatabricks
from langchain_core.messages import HumanMessage
from IPython.display import display, Markdown

def ask_databricks_llm(question: str, endpoint_name: str = "databricks-gpt-oss-120b"):
    """
    Connects to a Databricks Foundation Model using the LangChain framework.
    
    Args:
        question (str): The prompt to send to the model.
        endpoint_name (str): The Databricks serving endpoint to use.
    """
    
    # 1. Initialize the Chat Model
    # ChatDatabricks is a LangChain wrapper for Databricks Model Serving.
    # It automatically handles the connection and authentication.
    chat_model = ChatDatabricks(
        endpoint=endpoint_name,
        temperature=0.1 # Lower temperature = more factual/deterministic responses
    )

    try:
        # 2. Invoke the model
        # 'invoke' is the standard LangChain method to get a response.
        response = chat_model.invoke(question)
        
        # 3. Format and display the output
        # Markdown makes the LLM response (which often includes lists/code) 
        # look professional within the Databricks notebook.
        display(Markdown(f"### AI Response\n{response.content}"))
        
    except Exception as e:
        print(f"Error connecting to model: {e}")

# --- EXECUTION ---
if __name__ == "__main__":
    query = "What are the 3 key benefits of Delta Lake?"
    ask_databricks_llm(query)
```

```python
"""
This script demonstrates how to interact with Databricks Model Serving 
using the Foundation Model APIs. It utilizes the Databricks SDK to 
interface with LLMs hosted within the workspace.
"""

from databricks.sdk import WorkspaceClient

def get_completion(prompt: str, model_name: str = "databricks-gpt-oss-120b") -> str:
    """
    Sends a prompt to a Databricks-hosted Foundation Model and returns the response.

    Args:
        prompt (str): The user question or instruction.
        model_name (str): The name of the serving endpoint (found in the Serving UI).

    Returns:
        str: The text content of the model's response.
    """
    
    # 1. Initialize the WorkspaceClient
    # Best Practice: In a Databricks notebook, this automatically uses 
    # the existing login credentials of the current user.
    w = WorkspaceClient()

    # 2. Access the OpenAI-compatible client
    # Databricks provides an OpenAI-compatible interface for ease of use.
    client = w.serving_endpoints.get_open_ai_client()

    # 3. Request a completion from the model
    # We use a try-except block to handle potential API or connection errors gracefully.
    try:
        response = client.chat.completions.create(
            model=model_name,
            messages=[
                {"role": "system", "content": "You are a helpful Databricks assistant."},
                {"role": "user", "content": prompt}
            ],
            temperature=0.7  # Adding temperature helps control creativity vs. accuracy
        )
        
        # Extract and return the text content
        return response.choices[0].message.content

    except Exception as e:
        return f"An error occurred: {e}"

# --- EXECUTION ---
if __name__ == "__main__":
    # Define the query
    user_query = "How do I optimize a Delta table? Respond in 3-5 sentences"
    
    # Call the function and display the result
    result = get_completion(user_query)
    print(f"Model Response:\n{'-'*20}\n{result}")
```
