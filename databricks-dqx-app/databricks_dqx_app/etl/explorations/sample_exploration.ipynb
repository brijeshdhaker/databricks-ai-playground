### Example Exploratory Notebook

Use this notebook to explore the data generated by the pipeline in your preferred programming language.

**Note**: This notebook is not executed as part of the pipeline.

```python
# Add this line at the very top of your notebook to define the widget
dbutils.widgets.text("catalog", "workspace")
dbutils.widgets.text("schema", "default")
```

```python
try:

    catalog = dbutils.widgets.get("catalog")
    schema = dbutils.widgets.get("schema")

except Exception:
    # Fallback if widget doesn't exist
    catalog = "workspace" 
    schema  = "default" 

# Set default catalog and schema
spark.sql(f"USE CATALOG {catalog}")
spark.sql(f"USE SCHEMA {schema}")
```

!!! Before performing any data analysis, make sure to run the pipeline to materialize the sample datasets. 
The tables referenced in this notebook depend on that step.

##### Explorer data from sample_trips

```python
display(spark.sql(f"SELECT * FROM {catalog}.{schema}.sample_trips"))
```
